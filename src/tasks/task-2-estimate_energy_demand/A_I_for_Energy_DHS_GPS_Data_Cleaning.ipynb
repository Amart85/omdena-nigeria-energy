#### ‚úî Step 5: Look for outliers

Now we look for outliers in our dataset. We can do this using the `df.describe()` method.

```python
df1.describe()
```

After running the command, we see that all columns have a minimum value of `-9999.00`. This value seems too low of our data. For example, population counts cannot be a negative value and temperature values should not go as low as `-9999.0`.

We will have to deal with these. I chose to remove the rows containing these values from the dataset since we have a lot of data.

```python
df1[df1['Annual_Precipitation_2015'] == -9999.0]
```

```python
df1[df1['Annual_Precipitation_2015'] == -9999.0].value_counts()
```

```python
df1 = df1[df1['Annual_Precipitation_2015'] != -9999.0]

df1.head()
```
